```python
#mapping categorical variables with their subcategories to help with visualization analysis 
boomBikes['season']=boomBikes.season.map({1: 'spring', 2: 'summer',3:'fall', 4:'winter' })
boomBikes['mnth']=boomBikes.mnth.map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'})
boomBikes['weathersit']=boomBikes.weathersit.map({1: 'Clear',2:'Mist + Cloudy',3:'Light Snow',4:'Snow + Fog'})
boomBikes['weekday']=boomBikes.weekday.map({0:'Sun',1:'Mon',2:'Tue',3:'Wed',4:'Thu',5:'Fri',6:'Sat'})

boomBikes.head()
```

As we can see, there are several variables correlated to count variable (cnt) such as registered, casual, temp, atemp, yr, workingday, we can thereby conduct a linear regresssion model

```python
#dropping the not required variables from the dataset 
#dropping the variables atemp, instant, dteday, casual, registered as they are not needed for the analysis 
#atemp is not needed as temp is already being used, dteday and casual are also not required for regression analysis 

boomBikes = boomBikes.drop(['atemp', 'instant', 'dteday', 'casual', 'registered'], axis=1)
boomBikes.head()
```

```python
#creating dummy variables 
#creating dummy variables for the variables of month, weekday, weathersit, seasons
month = pd.get_dummies(boomBikes.mnth, drop_first=True)
weekday = pd.get_dummies(boomBikes.weekday, drop_first=True)
weathersit = pd.get_dummies(boomBikes.weathersit, drop_first=True)
season = pd.get_dummies(boomBikes.season, drop_first=True)
```

```python
#adding the dummy variables to the original dataframe
boomBikes = pd.concat([boomBikes,month, weekday, weathersit, season], axis=1)
boomBikes.head()
```

```python
# dropping the variables season,mnth,weekday,weathersit as we have created the dummies for it
boomBikes.drop(['season','mnth','weekday','weathersit'], axis = 1, inplace = True)
boomBikes.head()
```

# Step 2: Preparing the data for model training (train-test split, rescaling)

```python
#splitting the dataset into train and test sets
boomBikes_train, boomBikes_test = train_test_split(boomBikes, train_size=0.7, random_state=100)
```

```python
#checking the shape of the training dataset
boomBikes_train.shape
```

```python
#checking the shape of the test dataset
boomBikes_test.shape
```

```python
#we have to rescale the variables like hum, temp, windspeed, cnt as they have large values as compared to the other variables of the dataset
scaler = MinMaxScaler()
scaler_var = ['hum', 'windspeed', 'temp', 'cnt']
boomBikes_train[scaler_var] = scaler.fit_transform(boomBikes_train[scaler_var])
```

```python
#checking the normalized values of the train set after performing scaling 
boomBikes_train.describe()
```

We can see that the max amount is 1 in all cases of the variables which is an indication that our rescaling process has worked

```python
# checking the correlation coefficients to see which variables are highly correlated post data preparation and rescaling

plt.figure(figsize = (22, 12))
sns.heatmap(boomBikes_train.corr(), cmap="Greens", annot=True)
plt.title('Heatmap to check correlation after data preparation and rescaling')
plt.show()
```

As we can see from the heatmap, the count seems to be heavily correlated with the temp variable.

# Step 3: Training the model

```python
#building our first model using the variable temp
#preparing the variables for model building 

y_train = boomBikes_train.pop('cnt')
X_train = boomBikes_train
```

```python
#checking the variables
y_train.head()
```

```python
X_train.head()
```

```python
#add a constant (intercept)
X_train_sm = sm.add_constant(X_train['temp'])

#create first model
lr = sm.OLS(y_train, X_train_sm)

#fit
lr_model = lr.fit()

#params
lr_model.params
```

```python
#checking model summary 
lr_model.summary()
```

R-squared is 0.416 which means 41.6% of the variance in the count quantity is explained by temp

```python
#visualising our data with a scatter plot and the fitted regression line to see the best fit line
plt.scatter(X_train_sm.iloc[:, 1], y_train)
plt.plot(X_train_sm.iloc[:, 1], 0.1690 + 0.6409*X_train_sm.iloc[:, 1], 'r')
plt.title('Fitted regression line as explained by temp')
plt.show()
```

```python
#adding another variable thus performing multiple regression 
#adding variable yr and checking to see if it improves the R-squared
X_train_sm = X_train[['temp', 'yr']]
X_train_sm = sm.add_constant(X_train_sm)

#create second model
lr = sm.OLS(y_train, X_train_sm)

#fit
lr_model = lr.fit()

#params
lr_model.params
```

```python
#checking summary with temp and yr as selected variables
lr_model.summary()
```

R-squared is now 0.692 which means 69.2% of the variance in the count quantity is explained by temperature and year data

# Using RFE:
We have found out the R-squared values and the best fit lines using the manual approach for feature selection.
We will now using the automated approach for selecting the values required for building the most optimized 
models and dropping the values which are redundant for our model building approach. We will now use RFE for automated 
approach, along with VIF to determine the variables to drop.

```python
#checking the number of columns in our dataset
len(boomBikes.columns)
```

```python
#creating the RFE object
lm = LinearRegression()
lm.fit(X_train, y_train)

#setting feature selection variables to 15
rfe = RFE(lm, n_features_to_select = 15) 

#fitting rfe ofject on our training dataset
rfe = rfe.fit(X_train, y_train)
```

```python
#checking the elements selected and the ones rejected in a list after rfe
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
#getting the selected feature variables in one variable
true_rfe = X_train.columns[rfe.support_]
```

```python
#checking the values of true_rfe
true_rfe
```

```python
len(true_rfe)
```

```python
#building model using selected RFE variables
#creating training set with RFE selected variables
X_train_rfe = X_train[true_rfe]
```

```python
#adding constant to training variable
X_train_rfe = sm.add_constant(X_train_rfe)

#creating first training model with rfe selected variables
lr = sm.OLS(y_train, X_train_rfe)

#fit
lr_model = lr.fit()

#params
lr_model.params
```

```python
#summary of model
lr_model.summary()
```

```python
#checking the VIF of the model 

#dropping the constant variables from the dataset
X_train_rfe = X_train_rfe.drop(['const'], axis = 1)
```

```python
#calculating the VIF of the model
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```
